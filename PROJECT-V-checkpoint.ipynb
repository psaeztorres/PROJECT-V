```python
def get_1percent_data (path,name):
    df = pd.read_csv(path)
    df= df.sample(frac=0.001)
    df.to_csv(f"data/{name}.csv")
    name = pd.read_csv(f"data/{name}.csv")
    name = pd.DataFrame(name)
    return df 
```

```python
rd_2016_01=get_1percent_data ("starting_data\yellow_tripdata_2016-01.csv",'rd_2016_01')
```

```python
rd_2016_02=get_1percent_data ("starting_data\yellow_tripdata_2016-02.csv",'rd_2016_02')
```

```python
rd_2016_03=get_1percent_data ("starting_data\yellow_tripdata_2016-03.csv",'rd_2016_03')
```

```python
def full_join_df (*df):
    dfs = list(df)
    concatenated_df = pd.concat(dfs, ignore_index=True)
    return concatenated_df
```

```python
taxis = full_join_df (rd_2016_01,rd_2016_02,rd_2016_03)
taxis.head()
```

```python
def get_day_hour(df,column_name) : 
    df['day_hour_pickup'] = pd.to_datetime(df[column_name])
    df['day_hour_pickup'] =  df['day_hour_pickup'].dt.round('H')
    df['year'] = df['day_hour_pickup'].dt.year
    df['month'] = df['day_hour_pickup'].dt.month
    df['day'] = df['day_hour_pickup'].dt.day
    df['hour'] = df['day_hour_pickup'].dt.hour
    return df
```

```python
taxi = get_day_hour(taxis,'tpep_pickup_datetime')
```

```python
taxi.head()
```

```python
def trip_duration (df,pickup_time_column, dropoff_time_column):
    df['trip_duration(min)'] = (pd.to_datetime(df[dropoff_time_column])-pd.to_datetime(df[pickup_time_column])).dt.total_seconds() / 60
    return df
    
```

```python
trip_duration (taxi,'tpep_pickup_datetime', 'tpep_dropoff_datetime')
```

```python
taxi.to_csv("data/taxi.csv")
```

```python
def create_day_hour (df,date_column, time_column):
    df['dia_hora'] = df[date_column]+' '+df[time_column]
    df['dia_hora'] = pd.to_datetime(df['dia_hora'], format='%Y-%m-%d %I:%M %p').dt.round('H')
    return df
```

```python
weather = create_day_hour (weather,'date', 'TimeEST')
```
